```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
datacp1 = pd.read_csv('alldata_umap.csv')

```

```python
sns.set(font="Arial")

sns.set_style(rc = {'axes.facecolor': '#FFFFFF','axes.edgecolor': 'black','xtick.color': 'black',
 'ytick.color': 'black',
 'xtick.direction': 'in',
 'ytick.direction': 'in',
  'grid.linestyle': '-',
  'xtick.bottom': True,
 'xtick.top': False,
 'ytick.left': True,
 'ytick.right': False,
 'axes.spines.left': True,
 'axes.spines.bottom': True,
 'axes.spines.right': False,
 'axes.spines.top': False,
'axes.facecolor': 'white',
 'axes.edgecolor': 'black',
 'axes.grid': False,
 'axes.axisbelow': 'line',
 'axes.labelcolor': 'black',
 'figure.facecolor': 'white'})


sns_plot = sns.regplot(y="ApathySM", x="UMAP2",x_bins=10,data=datacp1,scatter_kws={'s':10, 'alpha':0.9}, 
            fit_reg = True, color = '#FFBF00')
#sns_plot.set_xlim(-10, 10)
#sns_plot.set_ylim(10,15)
sns_plot.set_xticklabels([str(i) for i in sns_plot.get_xticks()], fontsize= 24)
sns_plot.set_yticklabels([str(i) for i in sns_plot.get_yticks()], fontsize= 24)
sns_plot.set_xlabel('')      
sns_plot.set_ylabel('')  

scatter_fig1 = sns_plot.get_figure()
scatter_fig1.savefig('UMAP2_apathySM.pdf',dpi=1200)


```
